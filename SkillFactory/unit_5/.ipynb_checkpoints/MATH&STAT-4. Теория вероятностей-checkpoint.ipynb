### 4.2. Вероятность. Общие понятия
### Задание 4.2.1
Выберите верные утверждения о вероятностных пространствах  

Ответ:  
- Сумма вероятностей всех исходов должна быть равна единице
- При подбрасывании монетки «решка» является элементарным исходом
- В примере с эклерами элементарными исходами являются целые числа от 1 до 6

Корректна ли такая вероятностная модель?  

Ответ:  
Для решения нужно найти сумму вероятностей всех исходов.

### 4.4. Условная вероятность
### Задание 4.4.1
Выберите верные утверждения:
-  P (A|B) читается как A при условии B
- Условная вероятность равна нулю, если условие и интересующее нас событие не пересекаются  

### Задание 4.4.2
1) Среди всех проданных в ларьке цветов, 20% являются красными розами. Обозначим

A {«цветок является розой»}, B {«цветок красный»}.

Какая вероятность задана?

- Вероятность пересечения  

2) Среди всех проданных в ларьке красных цветов, 80% являются розами. Какая вероятность задана?

- Условная P (A|B)

3) Среди всех проданных в ларьке роз, 40% — красные. Какая вероятность задана?

- Условная P (B|A)
