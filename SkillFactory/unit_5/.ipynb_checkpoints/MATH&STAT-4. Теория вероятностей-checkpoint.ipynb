### 4.7. Задача на формулу полной вероятности


### Задача 5.7.1
2% клиентов испытали wow-effect от вашего сервиса и рассказывают о вас буквально всем. Из них 70% приведут новых клиентов. Еще 20% клиентов периодически оставляют вам положительные отзывы в соцсетях и на сайте. Из них 10% приведут новых клиентов. Что думают оставшиеся 78% клиентов, вы не знаете. Из них 4% приведут новых клиентов.

***Решение:***  
P(WOW)=2%  
P(CSC) = 20%  
P(Unknown) = 78%  

NC - New Client - приведет нового клиента  
P(NC|WOW)=70%  
P(NC|CSC)=10%  
P(NC|Unknown)=4%  

1. Выберите верное утверждение:  

***Ответ:*** P(«клиент приведет нового клиента» | «клиент испытал wow-effect») = P(NC|WOW) = 70%  

2. Если вы хотите найти вероятность, что случайный клиент вашего сервиса приведет нового клиента, что вы возьмёте за разбиение?  
***Определение:*** Разбиение — это набор взаимно-исключающих совместно исчерпывающих событий.  

***Ответ:*** типы клиентов (***Объяснение*** - потому что: $P(WOW \cap CSC) = P(WOW \cap Unknown) = P(Unknown \cap CSC) = 0 $, а также  
$P(WOW) + P(CSC) + P(Unknown) = 0.02 + 0.2 + 0.78 = 1$ )

3. Найдите вероятность, что случайный клиент вашего сервиса приведет нового клиента: 
$P(NC) = P(NC \cap WOW) + P(NC \cap CSC) + P(NC \cap Unknown) =$ 
$= P(NC|WOW)*P(WOW) + P(NC|CSC)*P(CSC) + P(NC|Unknown)*P(Unknown) = $
$= 0.7*0.02 + 0.1*0.2 + 0.04*0.78 = 0.0652$  

***Ответ:*** 0.0652

4. Какой из типов клиентов дает наибольший вклад в приведение новых клиентов?  
$P(WOW|NC) = \frac {P(NC \cap WOW)}{P(NC)} = \frac{0.7*0.02}{0.0652} = 0.2147$  
$P(CSC|NC) = \frac {P(NC \cap CSC)}{P(NC)} = \frac{0.1*0.2}{0.0652} = 0.3067$  
$P(Unknown|NC) = \frac {P(NC \cap Unknown)}{P(NC)} = \frac{0.04*0.78}{0.0652} = 0.4786$  

***Ответ:*** клиенты-тихони, про которых ничего не известно верно = $P(Unknown|NC) = 0.4786$  

5. Кто-то из ваших клиентов привел нового клиента. Какая вероятность, что это сделал клиент тихоня? (*Ответ округлите до сотых*)

***Ответ:*** $P(Unknown|NC) = 0.48$
