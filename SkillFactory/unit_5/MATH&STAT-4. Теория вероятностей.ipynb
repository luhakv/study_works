P(NC|WOW)=70%  
P(NC|CSC)=10%  
P(NC|Unknown)=4%

***Ответ:*** P(«клиент приведет нового клиента» | «клиент испытал wow-effect») = P(NC|WOW) = 70%

***Ответ:*** типы клиентов (***Объяснение*** - потому что: $P(WOW \cap CSC) = P(WOW \cap Unknown) = P(Unknown \cap CSC) = 0 $, а также  
$P(WOW) + P(CSC) + P(Unknown) = 0.02 + 0.2 + 0.78 = 1$ )

$P(NC) = P(NC \cap WOW) + P(NC \cap CSC) + P(NC \cap Unknown) =$ 
$= P(NC|WOW)*P(WOW) + P(NC|CSC)*P(CSC) + P(NC|Unknown)*P(Unknown) = $

$P(Unknown|NC) = \frac {P(NC \cap Unknown)}{P(NC)} = \frac{0.04*0.78}{0.0652} = 0.4786$

### 4.10. Независимость событий

В приложениях теории вероятностей к анализу данных очень важным является понятие независимости. Сейчас мы говорим о вероятности событий, поэтому и независимость рассмотрим тоже для событий.  
Определение:  События А и В называются независимыми, если вероятность их пересечения равна произведению вероятностей  

$P(A\cap B)=P(A)\cdot P(B)$  

Главное свойство независимых событий: Вероятность А не зависит от наличия условия. То есть при наступлении события В событие А происходит так же часто, как и без него  

$P(A|B)=P(A)$
