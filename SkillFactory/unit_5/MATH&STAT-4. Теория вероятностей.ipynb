### 4.2. Вероятность. Общие понятия
### Задание 4.2.1
Выберите верные утверждения о вероятностных пространствах  

Ответ:  
- Сумма вероятностей всех исходов должна быть равна единице
- При подбрасывании монетки «решка» является элементарным исходом
- В примере с эклерами элементарными исходами являются целые числа от 1 до 6

Корректна ли такая вероятностная модель?
