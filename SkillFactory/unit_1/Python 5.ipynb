```python
import pandas as pd #Укажите  возраст (Age) самого молодого игрока, у которого хладнокровие (Composure) и реакция (Reactions) превышают 90% от максимального значения, представленного в датасете.
football = pd.read_csv('data_sf.csv')

football[(football.Composure > football.Composure.max() * 0.9) & (football.Reactions >football.Reactions.max() * 0.9)].Age.min()
```

```python
import pandas as pd #насколько средняя реакция (Reactions) самых взрослых игроков (т.е. игроков, чей возраст (Age) равен максимальному) больше средней реакции самых молодых игроков
football = pd.read_csv('data_sf.csv')

football[(football[football.Age == football.Age.max()].Reactions.mean()) - (football[football.Age == football.Age.min()].Reactions.mean())
```

```python
import pandas as pd 
'''во сколько раз средняя сила удара (ShotPower) самых агрессивных игроков (игроков с максимальным значением 
показателя "Агрессивность" (Aggression)) выше средней силы удара игроков с минимальной агрессией.'''
football = pd.read_csv('data_sf.csv')

football[football.Aggression == football.Aggression.max()].ShotPower.mean()
football[football.Aggression == football.Aggression.min()].ShotPower.mean()
```

```python
import pandas as pd 
'''во сколько раз средняя зарплата (Wage) голкипера (Position, GK) с максимальным значением показателя " Рефлексы" (GKReflexes) 
выше средней зарплаты голкипера с максимальным значением показателя "Владение мячом" (GKHandling). 
Ответ округлите до сотых.'''
football = pd.read_csv('data_sf.csv')

#football[(football.Position == 'GK')&(football.GKReflexes == football.GKReflexes.max())].Wage.mean() 
#football[(football.GKHandling == football.GKHandling.max())].Wages.mean()
football[(football.Position == 'GK')&(football.GKReflexes == football.GKReflexes.max())].Wage.mean()
football[(football.Position == 'GK')&(football.GKHandling == football.GKHandling.max())].Wage.mean()

football[(football.Position == 'GK')&(football.GKReflexes == football.GKReflexes.max())].Wage.mean() / football[(football.Position == 'GK')&(football.GKHandling == football.GKHandling.max())].Wage.mean()
```

```python
import pandas as pd 
'''во сколько раз средняя сила удара (ShotPower) самых агрессивных игроков (игроков с максимальным значением 
показателя "Агрессивность" (Aggression)) выше средней силы удара игроков с минимальной агрессией.'''
football = pd.read_csv('data_sf.csv')

football[football.Aggression == football.Aggression.max()].ShotPower.mean() / football[football.Aggression == football.Aggression.min()].ShotPower.mean()
```

```python
import pandas as pd 
'''насколько средняя реакция (Reactions) самых взрослых игроков (т.е. игроков, чей возраст (Age) равен максимальному) 
больше средней реакции самых молодых игроков. '''
football = pd.read_csv('data_sf.csv')

football[football.Age == football.Age.max()].Reactions.mean() - football[football.Age == football.Age.min()].Reactions.mean()
```
